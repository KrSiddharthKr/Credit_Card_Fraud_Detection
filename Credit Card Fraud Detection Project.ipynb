# Importing the dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading the dataset to a Pandas DataFrame
credit_card_data = pd.read_csv('credit_data.csv')
```

```python
# print first 5 rows of the dataset
credit_card_data.head()
```

```python
# print last 5 rows of the dataset
credit_card_data.tail()
```

```python
# dataset information
credit_card_data.info()

# null values means missing values and non-null values means that valuess are present 
```

```python
# checking the number of missing values in each column
credit_card_data.isnull().sum()
```

```python
# distribution of legit transaction and fraudulent transaction
credit_card_data['Class'].value_counts()
```

# This dataset is highly unbalanced

## 0 -> Legit transaction
## 1 -> Fraudulent transaction

```python
# separating the data for analysis

legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measures of the data

legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both transactions

credit_card_data.groupby('Class').mean() # it will compare the mean value of all the columns for these two classes : legit(0) and fraudulent(1) transaction
```

## Under-Sampling 

### Build a sample dataset containing similar distribution of normal transaction and Fraudulent transaction 

### Number of Fraudulent Transaction  --> 492

```python
legit_sample = legit.sample(n = 492) # random datapoints from legit
```

## Concatenating two DataFrames

```python
new_dataset = pd.concat([legit_sample , fraud] , axis = 0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean() # this particular step is important to check whether we got a good sample or a bad sample
# if bad sample then mean values will be very different
```

## Splitting the data into Features and Targets

```python
X = new_dataset.drop(columns = 'Class' , axis = 1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

## Split the data into Training data and Testing data 

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify = Y, random_state = 2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

## Model Training 

## Logistic Regression

```python
model = LogisticRegression()
```

```python
# training the Logistic Regression model with Training data
model.fit(X_train, Y_train)
```

## Model Evaluation

## Accuracy Score

```python
# accuracy on training data

X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on training data : ', training_data_accuracy)
```

```python
# accuracy on test data

X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on test data : ', test_data_accuracy)
```
